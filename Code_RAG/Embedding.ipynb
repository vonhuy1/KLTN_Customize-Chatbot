```python
from langchain_community.embeddings import HuggingFaceBgeEmbeddings

model_name = "BAAI/bge-large-en-v1.5"
model_kwargs = {"device": "cpu"}
encode_kwargs = {"normalize_embeddings": True}
```

```python
embeddings_cohere = CohereEmbeddings(model="embed-english-light-v3.0")
embedding_model = IpexLLMBgeEmbeddings(
    model_name="BAAI/bge-large-en-v1.5",
    model_kwargs={},
    encode_kwargs={"normalize_embeddings": True},
)
embeddings_google = GoogleGenerativeAIEmbeddings(model="models/embedding-001")
hf_embedding = HuggingFaceBgeEmbeddings(
    model_name=model_name, model_kwargs=model_kwargs, encode_kwargs=encode_kwargs
)
embeddings_fake = FakeEmbeddings(size=4000)

```

```python
import time
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
from langchain.document_loaders import ObsidianLoader
from langchain.document_loaders import PDFMinerLoader
from langchain.document_loaders import PDFPlumberLoader
from langchain.document_loaders import PyPDFium2Loader
from langchain.document_loaders import UnstructuredPDFLoader
from langchain.document_loaders import PyMuPDFLoader
```

```python
loader = UnstructuredPDFLoader("./example_data/demo1.pdf")
pages_PdfUnstructured = loader.load()
```

```python
text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(chunk_size=5000, chunk_overlap=3000)
texts = text_splitter.split_documents(pages_PdfUnstructured)
```

```python
from langchain.vectorstores import Chroma
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.chains import RetrievalQA
#from langchain.document_loaders import TextLoader
from langchain_community.document_loaders import PyPDFLoader
from langchain.document_loaders import UnstructuredPDFLoader
import pathlib
import textwrap
import google.generativeai as genai
import PIL.Image
import urllib.request 
from PIL import Image
from IPython.display import display
from IPython.display import Markdown
```

```python
start_time = time.time()
persist_directory = 'db1'
vectordb1_embd_hf = Chroma.from_documents(documents=texts,
                                 embedding=hf_embedding ,
                                 persist_directory=persist_directory)
end_time = time.time()
elapsed_time = end_time - start_time
print(f"Thời gian thực hiện: {elapsed_time:.2f} giây")
```

```python

start_time=time.time()
persist_directory = 'db2'
vectordb1_embd_fake = Chroma.from_documents(documents=texts,
                                 embedding=embeddings_fake,
                                 persist_directory=persist_directory)
end_time = time.time()
elapsed_time = end_time - start_time
print(f"Thời gian thực hiện: {elapsed_time:.2f} giây")
```

```python

start_time=time.time()
persist_directory = 'db3'
vectordb1_embd_bge = Chroma.from_documents(documents=texts,
                                 embedding=embedding_model,
                                 persist_directory=persist_directory)
end_time = time.time()
elapsed_time = end_time - start_time
print(f"Thời gian thực hiện: {elapsed_time:.2f} giây")
```

```python
start_time = time.time()
persist_directory = 'db4'
vectordb_embd_google = Chroma.from_documents(documents=texts,
                                 embedding=embeddings_google,
                                 persist_directory=persist_directory)
end_time = time.time()
elapsed_time = end_time - start_time
print(f"Thời gian thực hiện: {elapsed_time:.2f} giây")
```
